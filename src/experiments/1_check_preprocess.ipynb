# Experiment 2: Inspect Preprocessed Images
This experiment is to check if the image is augmented in the way we want it to.

```python
# Import section
import os
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from PIL import Image
```

```python
DATA_PATH = os.path.join(os.getcwd(), '..', 'data', 'processed', 'TrainVal')
```

```python
file_name = 'german_shorthaired_126.jpg'
```

```python
mask_name = file_name.replace('.jpg', '.png')
aug_0 = 'augmented_0_' + file_name
aug_1 = 'augmented_1_' + file_name
aug_2 = 'augmented_2_' + file_name
aug_3 = 'augmented_3_' + file_name
actual = 'resized_' + file_name

mas_0 = 'augmented_0_' + mask_name
mas_1 = 'augmented_1_' + mask_name
mas_2 = 'augmented_2_' + mask_name
mas_3 = 'augmented_3_' + mask_name
mas_actual = 'resized_' + mask_name

file_names = [aug_0, aug_1, aug_2, aug_3, actual]
mask_names = [mas_0, mas_1, mas_2, mas_3, mas_actual]
```

```python
def plot_images(image_filenames, dir="color"):
    plt.figure(figsize=(20, 5))  # Adjust figure size as needed


    # Loop through image filenames and plot them
    for idx, filename in enumerate(image_filenames):
        image_path = os.path.join(DATA_PATH, dir, filename)

        if os.path.exists(image_path):  # Avoid errors if file is missing
            plt.subplot(1, 5, idx + 1)
            plt.imshow(Image.open(image_path))
            plt.axis("off")  # Hide axis
            plt.title(f"Image {idx}")  # Add title for clarity

    # Show the final visualization
    plt.tight_layout()
    plt.show()
```

```python
plot_images(file_names)
```

```python
plot_images(mask_names, dir="label")
```
